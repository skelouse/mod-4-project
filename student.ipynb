```python
stop_list = stopwords.words('english')
stop_list += string.punctuation
```

```python
def encode(series, stopwords=stop_list):
    """
    Encodes a series into a bag of words series, removing stop words and URLS
    
    Parameters
    -----------------------------
    series[pd.Series]:
        A series of strings to be converted
    stop_list[list]:
        A list of stop words to be removed from the Series
    
    Returns
    -----------------------------
    series[pd.Series]:
        where "@Wesley83 3g iphone 3 hrs Tweeting http://google.com"
        = [wesley83, 3g, iphone, 3, hrs, tweeting]
        """
    series = series.replace('(http[^ ]*)|(www\.[^ ]*)', '', regex=True)
    series = series.str.replace("[^\w\s]", "").str.lower().copy()
    series = series.apply(lambda x: [i for i in x.split() if i not in stop_list])
    return series
```

### Functions

```python
def cust_classification_report(estimator, X, y, X_test, y_test,
                               classes=['Negative',
                                        'Positive'],
                               return_axes=False):
    """
    Scores the model, and plots a confusion matrix for the train and test
    data
    
    Parameters
    ----------------------------------------
    estimator[sklearn.Classfier]::
        The estimator that the classfication report is made for.
    X[Series or Matrix]::
        The features to predict with that the estimator was trained on.
    y[Series or Matrix]::
        The labels that will be predicted that the estimator was trained on.
    X_test[Series or Matrix]::
        The test features to predict with.
    y_test[Series or Matrix]::
        The test labels that will be predicted.
    classes[list]::
        A list of classes for the labels on the confusion matrices
    return_axes[bool]:: Default (False)
        Whether or not to return the fig and axes
    
    Returns if return_axes
    ----------------------------------------
    fig, axes
        The figure, and two axes of the confusion matrices
    
    """
    print("Training score:\n", estimator.score(X, y))
    print("="*20)
    print("Testing score:\n", estimator.score(X_test, y_test))
    print("="*20)
    fig, (ax1, ax2) = plt.subplots(figsize=(15, 6), ncols=2)
    ax1.set_title("Training data")
    plot_confusion_matrix(estimator, X, y, ax=ax1, cmap='Blues',
                          normalize='true', display_labels=classes)
    ax1.set_yticklabels(ax1.get_yticklabels(), rotation=90)
    ax2.set_title("Testing data")
    plot_confusion_matrix(estimator, X_test, y_test, ax=ax2, cmap='Blues',
                          normalize='true', display_labels=classes)
    ax2.set_yticklabels(ax2.get_yticklabels(), rotation=90)
    if return_axes:
        return fig, (ax1, ax2)
```

```python
def grid_search(estimator, X, y, X_test, y_test,
                params={}, scoring='accuracy', verbose=True):
    """
    Models using GridSearchCV to find the best params for a specific model

    Parameters
    ----------------------------------------
    estimator[sklearn.Classfier]::
        The estimator that the classfication report is made for.
    X[Series or Matrix]::
        The features to predict with that the estimator was trained on.
    y[Series or Matrix]::
        The labels that will be predicted that the estimator was trained on.
    X_test[Series or Matrix]::
        The test features to predict with.
    y_test[Series or Matrix]::
        The test labels that will be predicted.
    params[dict]::
        parameters to test the model on
    scoring[str]::
        the mode of scoring to use see:
        https://scikit-learn.org/stable/modules/model_evaluation.html
    verbose[bool]::
        Whether to print out the classification report of the best model,
        and the params used    
    
    Returns
    ----------------------------------------
    clf_best[]::
        The best model that was found from the search
    """
    clf = GridSearchCV(estimator, params, scoring=scoring,
                       n_jobs=-1, cv=5)
    clf.fit(X, y)
    clf_best = clf.best_estimator_
    clf_best.fit(X, y)
    if verbose:
        print("Estimator:")
        print(clf.best_estimator_)
        print("="*20)
        cust_classification_report(clf_best, X, y, X_test, y_test)
    return clf_best
```

```python
def evaluate_nn(model, history, X_train, y_train, X_test, y_test,
                classes=["Negative", "Positive"]):


    def make_confusion(y_true, y_pred, ax):
        cm = confusion_matrix(y_true, y_pred, normalize='true')
        cm_df = pd.DataFrame(cm,index=classes,columns=classes)
        sns.heatmap(cm_df, annot=True, cmap='Blues', ax=ax)
    
    
    def decorate_print(x):
        print('', "="*80,'\n',
              ' '*20, '   %s:\n'%x,
              '='*80)
    y_train_pred = model.predict(X_train).flatten().round()
    y_test_pred = model.predict(X_test).flatten().round()
    decorate_print("Classification Report (Test Data)")
    report = classification_report(y_test, y_test_pred, target_names=classes)
    report = '\n' + report
    print(report.replace('\n', '\n' + ' '*20))
    decorate_print("Confusion Matrices")
    fig, (ax1, ax2) = plt.subplots(figsize=(15, 6), ncols=2)
    
    # Train data confusion matrix
    make_confusion(y_train, y_train_pred, ax1)
    ax1.set_title("Training data")

    # Test data confusion matrix
    make_confusion(y_test, y_test_pred, ax2)
    ax2.set_title("Testing data")

    plt.show()
    decorate_print("History of loss and accuracy graphs")
    
    df = pd.DataFrame(history.history)
    
    fig, (ax3, ax4) = plt.subplots(figsize=(15, 6), ncols=2)

    df[[col for col in df.columns if 'acc' in col]].plot(ax=ax3)
    ax3.set_title("Accuracy")
    
    df[[col for col in df.columns if 'loss' in col]].plot(ax=ax4)
    ax4.set_title("Loss")
```
